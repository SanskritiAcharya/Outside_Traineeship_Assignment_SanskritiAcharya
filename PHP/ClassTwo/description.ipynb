# Roman Calculator 

### How it works

1. User enters a number in input form.
2. The number is sent to PHP using POST method.
3. PHP checks if the number is between 1 and 3999.
4. If valid, an object of RomanCalculate class is created.
5. The number is stored inside class.
6. The program converts the number into Roman numerals step by step.
7. Result is displayed on the screen.
8. If invalid input is given, an error message is shown.

---

### OOP Concepts Used

- Class: RomanCalculate
- Object: Created using new
- Encapsulation: private variable used to store number
- Methods: putValue() and makeRoman().
